```python
from pydataset import data  #dataset library
```

```python
mtcars = data('mtcars')
```

```python
mtcarsDF2 = mtcars.copy()
```

```python
mtcarsDF2.head()
```

```python
#dfply u
```

```python
from dfply import *
```

```python
mtcarsDF2['carname'] = mtcarsDF2.index  #copy car names as column
```

```python
catcol = ['cyl', 'vs', 'am', 'gear', 'carb']
```

```python
mtcarsDF2.dtypes
```

```python
mtcarsDF2[catcol] = mtcarsDF2[catcol].astype('category')
```

```python
mtcarsDF2.dtypes
```

### start using dfply 

```python
mtcarsDF2 >> select(~X.mpg)
```

```python
mtcarsDF2 >> select(~X.mpg, ~X.wt) >> tail()  #exclude mpg, wt
```

```python
mtcarsDF2 >> select(X.mpg, X.gear, X.wt)  >> arrange(X.gear, -X.wt) >> head()
```

```python
mtcarsDF2 >> group_by(X.cyl, X.gear) >> summarise(meanMPG = X.mpg.mean())
```
