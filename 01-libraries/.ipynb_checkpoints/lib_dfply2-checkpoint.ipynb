# dfply library
https://github.com/kieferk/dfply

```python
#pip install dfply
from dfply import *
import pandas as pd
import numpy as np
from pydataset import data
```

```python
DF = data('mtcars')
DF.head()
```

```python
DF >> select(X.mpg, X.wt)  >> head()
```

```python
DF >> select(~X.mpg) >> head()
```

```python
DF >> drop(X.mpg)  >> head()
```

```python
DF >> mask(X.mpg > 25)  >> head()
```

```python
DF >> arrange(X.am, X.mpg) >> select(X.carb, X.mpg) >> head(10)
```

```python
DF >> mutate(newMPG = X.mpg * 1.4) >> head()
```

```python
DF >> group_by(X.gear, X.am)  >> summarize(mpgMean = X.mpg.mean(), wtMean = X.wt.max())
```

```python
DF >> mask(X.mpg > 25, X.am ==1)
```
