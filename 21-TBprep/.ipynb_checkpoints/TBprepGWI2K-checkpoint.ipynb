## Months to Years
groupby [Year].[Country/Region]  : Find Sum of CPI and countr rows

```python
DCI_agg = DCI_long.groupby(['Year','Country/Region'], as_index = False).agg( 
    sumCPI = ('CPI','sum'), countCPI = ('CPI','count')).reset_index()
```

```python
DCI_agg.head(2)
```

```python
type(DCI_agg)
```

```python
DCI_agg.countCPI.value_counts()
```

```python
DCI_agg['Year'] = DCI_agg['Year'].str.replace("Y","")  #remove Y
```
