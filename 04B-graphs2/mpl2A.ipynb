```python
# create a simple data and plot
y = [1,2,5,6,7]
x = [1,2,3,4,5]
plt.scatter(x=x, y=y)
plt.show(); #this will now show any output message
```

```python
plt.scatter(x,y, c='r')
plt.figure(figsize=(2,1))  # this applicable to next figure
plt.plot(x,y)
```

```python
plt.figure(figsize=(3,2))  # applicabel to both
plt.scatter(x,y, c='r')
plt.plot(x,y)
plt.text(3,4,'Point')
plt.show();
```

```python
plt.figure(figsize=(8,3))  # applicable to both
plt.scatter(x,y, c='r')
plt.plot(x,y)
for x1,y1 in zip(x,y):
    label1 = y1
    label2 = f"({x1},{y1})"
    plt.annotate(label2, (x1,y1), textcoords="offset points", xytext=(.2,.3), ha='left', va ='bottom')
plt.show();
```

```python
#bar
plt.figure(figsize=(8,3))  # applicable to both
plt.bar(x,y, color='r')
plt.plot(x,y)
for x1,y1 in zip(x,y):
    label = y1
    plt.annotate(label, (x1,y1), textcoords="offset points", xytext=(.2,.3), ha='left', va ='bottom')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show();
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4), sharey=False)
ax1.scatter(x1, y1)
ax2.plot(x1, y2) 

for a,b in zip(x1,y1):
    label1 = f"({a},{b})"
    ax1.annotate(label1, (a,b))
ax1.set_title('Sub Plot1 Title ')
#annotate in ax1
for a,b in zip(x1,y2):
    label2 = f"({a},{b})"
    ax2.annotate(label2, (a,b))
    ax2.title.set_text('Sub Plot 2 Title')  #another way
ax2.annotate(' Annotate Text Uniform Random Data', (4,90), color='red', size=12)
plt.suptitle('Plot Title : 2 sub Plots')  #common top title to subplots
fig.tight_layout()
plt.show();
```

```python
# using loop way
fig, ax = plt.subplots(2, 2, figsize=(8,4)) # 2 rows 2 cols
titles = ['Plot1', 'Plot2', 'Plot3', 'Plot4']
colors = ['r','g','b','y']
yvalues = [x3*1, x3*2, x3*3, x3*4]
LL=[]
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.plot(x3, yvalues[i], color=colors[i])
    plt.gca().set_title(titles[i])  #find the reference of current axes of subplot and set the title
plt.suptitle('Common Title of 4 plots', ha='left', fontsize=13)
plt.subplots_adjust(right=.9)
fig.legend(labels=titles)
plt.show()
```

```python
# using loop
plt.figure(figsize=(10,6)) 
for i in range(1,5):  #range to from 1
    plt.subplot(2, 2, i)
    plt.title('Scatter Plot of {}'.format(str(i)))
    plt.plot(x3, yvalues[i-1], color=colors[i-1])
    print(titles[i-1], colors[i-1])
    plt.legend(labels=titles[i-1], loc='upper right', title='Plot Name', fancybox=True)
plt.suptitle('Common Title of 4 plots', ha='left', fontsize=13)
#fig.legend(labels=titles)
plt.show()
```
