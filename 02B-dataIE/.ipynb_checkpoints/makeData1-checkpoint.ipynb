```python
import numpy as np
from sklearn.datasets import make_classification
import sys
```

```python
np.set_printoptions(threshold=sys.maxsize)
```

```python
np.set_printoptions(edgeitems=30, linewidth=1000,  formatter=dict(float=lambda x: "%.2g" % x))
```

```python
X, y = make_classification(1000, 20, n_informative=10, random_state=0)
X = X.astype(np.float32)
y = y.astype(np.int64)
```

```python
print(len(X), X)
```

```python
print(len(y), y)
```

```python
np.concatenate([X, y.reshape(-1,1)], axis=1)
```

```python
np.vstack((X, y.reshape(-1,1)))
```

```python
np.column_stack((X, y))
```
