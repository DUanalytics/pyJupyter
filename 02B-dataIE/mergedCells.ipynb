By default, pandas will read in the top row as the sole header row. You can pass a header argument into pandas.read_excel() that indicates how many rows are to be used as headers. In this case, we want header=[0, 1], indicating the first two rows. You might also have multiple sheets, so you can pass sheetname=None as well (this tells it to go through all sheets). The command would be:

```python
df2 = pd.read_excel(folder1 + file1, sheet_name ='S2')
df2.shape
```

```python
df2.head()
```

```python
df2B = pd.read_excel(folder1 + file1, sheet_name ='S2', header = [0,1])
df2B.shape
```
