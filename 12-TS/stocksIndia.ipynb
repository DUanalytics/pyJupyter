```python
# specify start and end dates
start1 = datetime(2022, 1, 1)
end1 = datetime(2022, 7, 10)
icici = data.DataReader('ICICIBANK.NS', start=start1, end=end1, data_source='yahoo')
icici.apply(lambda x: pd.concat([x.head(3), x.tail(2)]))   #head and tail together
```

```python
#Visualise
```

```python
stockData = icici['Close']  # select closing price
stockData
```

```python
#quick plot
stockData.plot()
plt.show();
```

```python
stockData.resample('BA').mean()
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.resample.html
#method for frequency conversion and resampling of time series. 
#The object must have a datetime-like index 
#(DatetimeIndex, PeriodIndex, or TimedeltaIndex),
#or the caller must pass the label of a datetime-like 
#series/index to the on/level keyword parameter.
```

```python
# Plot all types of values
#Resampling- aggregation, asfreq - data selection
stockData.plot(alpha=0.5, style='-')
stockData.resample('BA').mean().plot(style=':')
pd.options.display.float_format = '{:.2f}'.format
plt.figure(figsize=(10,6))
stockData.plot(alpha=0.5, style='-')
stockData.resample('BA').mean().plot(style=':')
stockData.asfreq('BA').plot(style='--');
plt.legend(['CLOSE','RESAMPLE','ASFREQ'], loc='upper right')
plt.show();
```

```python
#resample - avg of year, asfreq - value at end of year
#asfreq options of ffill, bfill 
#transactions only on business days
print(stockData[0:31])  # on non working days, there is no value.
#let us fill it with some value, eg 8, 9 Nov
data1 = stockData[0:31]  # select only first 31 rows
```

```python
#plot
data1.asfreq('D').plot(figsize=(10,5))
plt.show(); #break
```

```python
data1.asfreq('D', method='ffill') # now data has been filled with previous values
#7 Nov values (793) into 8 & 9th Nov
```

```python
data1.asfreq('D', method='ffill').plot(figsize=(10,5))
plt.show(); # continuous graphs
```

```python
#others bfill : plot together
fig, ax = plt.subplots(2, sharex=True, figsize=(10,5))
data1.asfreq('D').plot(ax=ax[0], marker='x')
data1.asfreq('D', method='ffill').plot(ax=ax[1], marker='D')
plt.show();
```

```python
#https://matplotlib.org/api/markers_api.html
# Back bill : 10 Nov values for 8 & 9th Nov (810)
data1.asfreq('D', method='bfill')
```

```python
fig, ax = plt.subplots(2, sharex=True, figsize=(10,8))
#data1.asfreq('D').plot(ax=ax[0], marker='x')
data1.asfreq('D').plot(ax=ax[0], marker='x')
data1.asfreq('D', method='ffill').plot(ax=ax[1], marker='D')
data1.asfreq('D', method='bfill').plot(ax=ax[1], marker='d')
ax[0].legend(['Orginal'])
ax[1].legend(['FORWARD fill','BACKWARD fill'])
plt.show();
```

```python
#drawing line - single and multiple
#print(data1)
#start1 = datetime(2022, 1, 1), end1 = datetime(2022, 2, 15)
data1.asfreq('D').plot(figsize=(10,5))
xcoords = [pd.to_datetime('2022-01-10'), pd.to_datetime('2022-02-01')]
for xc in xcoords:   plt.axvline(x=xc, color='b')
plt.show();
```

```python
# different line colors and back fill
colors = ['r','k']
data1.asfreq('D', method='bfill').plot(figsize=(10,5))
for xc,c in zip(xcoords,colors):
    plt.axvline(x=xc, label='line at x = {}'.format(xc), c=c)
plt.legend()
plt.show();
```
