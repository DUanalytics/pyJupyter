## Data Quality
Missing Values: It is possible that when the census was taken, few people refused to give information about their occupation/work-class. So, we would not have any kind of data for that column for those people. A good practice is to replace the missing values by median of the column or we could just drop these rows if they are not huge in number.
Outliers: It could also be that there was a person whose capital gain was $100,000 while the median capital gain was 0 dollars. These people are called outliers. It is a good practice to remove outliers from your training data, as these tend to shift your model towards the wrong results.
Duplicate data: We could also have information about the same person repeated in the data. We should remove all the duplicate data, as it could lead to overfitting if found in huge numbers.

## Correlation between Categorical columns
or categorical attributes, we can use something called the cross-tabulation, 
but before we look into it, it would be better and 
easier for us if we group the categorical attributes into 
lesser number of categories. For example, I could group the education values
to Dropout, HighSchoolGrad, Community College, Bachelors, Masters, Doctorate.

```python
#Grouping Education
df4.education.value_counts()
```

```python
#replace
df4['education'].replace('Preschool', 'Dropout', inplace=True)
df4.education.value_counts()
```

```python
df4[['education']] = df4[['education']].replace(dict.fromkeys(['10th','12th','1st-4th','5th-6th', '7th-8th' , '9th', '11th' ], 'Dropout'))
df4.education.value_counts()
```

```python
df4[['education']] = df4[['education']].replace(dict.fromkeys(['HS-grad'], 'HighSchoolGrad'))
df4[['education']] = df4[['education']].replace(dict.fromkeys(['HS-grad','Some-college',
                    'Assoc-acdm', 'Assoc-voc'], 'CommunityCollege'))
df4[['education']] = df4[['education']].replace(dict.fromkeys(['Prof-school'], 'Masters'))
df4.education.value_counts()
```

```python
#plot these education levels
fig = plt.figure(figsize = (20,5))
sns.countplot(x='education', data=df4)
plt.show();
```

```python
# Country Plot
df4['native-country'] = df4['native-country'].apply(lambda el: "United-States" if el == "United-States" else "Others")
df4['native-country'].value_counts()
```

```python
fig = plt.figure(figsize=(4,5))
sns.countplot(x='native-country', data=df4)
plt.show();
```

```python
#cross tab
pd.crosstab(df4['native-country'], df4['sex'], margins=True)
#ratio of men and women from US as well as other natives are more or less similar, 
#however, the number of US nationals are considerably more in the dataset.
```

```python
# Age
df4['age_bin'] = pd.cut(df4['age'], 10)
df4['age_bin'].value_counts()
```

```python
fig = plt.figure(figsize = (15,5))
sns.countplot(x='age_bin', hue='income', data = df4)
plt.show();
#plot shows that experience matters to earn more income. 
#More experienced people earn more than less experience people. This is true in the real world too.
```

```python
#Work Class
df4[['work-class']].value_counts()
```

```python
df4[['work-class']] = df4[['work-class']].replace(dict.fromkeys(['State-gov', 'Federal-gov', 'Local-gov'], 'Government'))
df4[['work-class']] = df4[['work-class']].replace(dict.fromkeys(['Self-emp-inc', 'Self-emp-not-inc'], 'Self-Employed'))
df4[['work-class']] = df4[['work-class']].replace(dict.fromkeys(['Without-pay', 'Never-worked'], 'Others'))
df4[['work-class']] = df4[['work-class']].replace(dict.fromkeys(['Private'], 'Private'))
df4[['work-class']].value_counts()
```

```python
fit = plt.figure(figsize = (20,5))
sns.countplot(x ='work-class', hue='income', data =df4)
plt.show();
# self-employed people in general make more money (in % /proportion)
```

```python
# education vs income
fit = plt.figure(figsize = (20,5))
sns.countplot(x ='education', hue='income', data =df4)
plt.show();
#people with higher education have a greater income.
```

```python
# Box Plot
fit = plt.figure(figsize = (12,6))
sns.boxplot(x ='work-class', y ='age', hue='sex', data =df4)
plt.show();
#The cross tabulation values show that more number of men are self employed rather than women. The box plot gives a few interesting observations:
#People who have never worked have a very young age. Women have a slightly higher median age than men for this working-class.
```

```python
# Box plot between race and age for different sex
plt.figure(figsize=(20,6))
sns.boxplot(x="race",y="age", hue="sex", data=df4)
plt.show();
#Men who work without pay have a median age of around 20-30, whereas women who work without pay have a median age of around 60-65.
#Most of the working class have a median age of around 40-50.
#People from "other" race have a younger median age than the rest of the races.
```
