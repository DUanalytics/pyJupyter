## Sometimes it is preferable to use numpy functions instead of pandas as there are faster
mean values from pandas and numpy
np.mean, np.min  - are from numpy https://numpy.org/doc/stable/reference/generated/numpy.std.html
min, max, std - are from pandas https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.std.html


```python
mtcarsDF.gear.mean()
```

```python
mtcarsDF.groupby(['gear'])[['mpg','wt']].mean() #pandas
```

```python
mtcarsDF.groupby(['gear'])[['mpg','wt']].agg(np.mean) #numpy 
```

```python
mtcarsDF.groupby(['gear']).aggregate({'mpg':[np.mean, 'mean',np.min, 'min',np.max,'max', np.std, 'std']})
```

```python
%timeit max(range(100000))
```

```python
%timeit mtcarsDF.aggregate({'mpg':[np.mean]})
```

```python
%timeit mtcarsDF.mpg.mean()
```
