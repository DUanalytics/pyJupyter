```python
# import the library
import folium
# Make an empty map
m = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=1)
# Show the map
m
```

```python
# Import the pandas library
import pandas as pd

# Make a data frame with dots to show on the map
data = pd.DataFrame({
   'lon':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5, 77.10, 92 ],
   'lat':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97, 28.7, 19],
   'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador',  'Delhi', 'Mumbai'],
   'value':[10, 12, 40, 70, 23, 43, 100, 43, 50, 67]
}, dtype=str)

data
```

```python
for i in range(0,len(data)):   print(data.iloc[i]['lat'], data.iloc[i]['lon'], data.iloc[i]['name'])
```

28.7041° N, 77.1025° E Delhi

```python
# add marker one by one on the map
for i in range(0,len(data)):   folium.Marker(  location=[data.iloc[i]['lat'], data.iloc[i]['lon']],  popup=data.iloc[i]['name'],).add_to(m)
# Show the map again
m
```

```python
# Make an empty map
n = folium.Map(location=[20,0], tiles="OpenStreetMap", zoom_start=2)

# add marker one by one on the map
for i in range(0,len(data)):
   folium.Marker(
      location=[data.iloc[i]['lat'], data.iloc[i]['lon']],
      popup=data.iloc[i]['name'],
      icon=folium.DivIcon(html=f"""<div style="font-family: courier new; color: blue">{data.iloc[i]['name']}</div>""")
   ).add_to(n)

# Show the map again
n
```
