# Stem Plot
https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.axes.Axes.stem.html
- A stem plot plots vertical lines at each x location from the baseline to y, and places a marker there.
- https://matplotlib.org/stable/gallery/lines_bars_and_markers/timeline.html
- https://python-charts.com/ranking/stem-plot-matplotlib/

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
locs = [0, 1, 2, 3, 4, 5, 6, 7, 8]
heads = [3, 4, 8, 2, 6, 5, 9, 4, 2]
```

```python
head_string = map(str, heads)
print(list(head_string))
print([str(x) for x in heads])
```

```python
#draw step plot
plt.stem(locs, heads, linefmt = 'red', markerfmt ='Dr', basefmt ='k:')
plt.show();
# change the color of the lines you can pass a new color to the linefmt argument,
#markerfmt argument allows customizing the color and shape of the markers of the lollipop.
#In the example below we are setting red (r) diamonds (D) as marker.
#base line color and style can be customized the same way as the other lines but using the basefmt argument. k+
```

```python
#step plot by group
n = np.arange(0, 10)

x1 = np.sin(n)
x2 = np.cos(n)

fig, ax = plt.subplots()

ax.stem(n, x1, 'b', markerfmt='bo', label="First",orientation='horizontal')
ax.stem(n, x2, 'g', markerfmt='go', label="Second")  #Different group in vertical 
#x = np.linspace(*ax.get_xlim())
#ax.plot(x, x*0, 'r-')
ax.legend()

plt.show();
#lollipop chart by group in Python just writing a new line. 
#In order to add a legend set the label with the label argument and use the legend function.
```

```python
# Data
x = np.arange(2, 12, 1)
y = x ** 2
fig, ax = plt.subplots()
ax.stem(x, y, bottom = 60)
plt.show()
# base line in red is set to zero by default, but you can override its position with the bottom argument.
```

```python
x = np.linspace(0.1, 2 * np.pi, 41)
y = np.exp(np.sin(x))

plt.stem(x, y, linefmt='--', use_line_collection=True, label=['xx'])
plt.show()
```

use_line_collection : bool, optional, default: False
If True, store and plot the stem lines as a LineCollection instead of individual lines. This significantly increases performance, and will become the default option in Matplotlib 3.3.
