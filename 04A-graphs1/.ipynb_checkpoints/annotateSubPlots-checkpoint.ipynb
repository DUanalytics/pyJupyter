```python
#subplots
fig = plt.figure(figsize=(8,8))   #  figure size (optional)
plt.subplot(a,b,c)                #  subplot 1
plt.plot(data_1)                  #  data 1
plt.subplot(a,b,c)                #  subplot 2
plt.plot(data_2)                  #  data 2
plt.tight_layout()                #  no overlap (optional)
plt.show()                        #  draw subplot group
-----------------------------------------------------------
where:
    a  - number of rows
    b  - number of columns
    c  - plot counter

-Three Rules for Subplots:
    - row (a) and column (b) numbers are constant per subplot group
    - plot counter (c) is initialized (=1) and then incremented for each subplot
    - subplot(abc) and subplot(a,b,c) are the same commands
```

## Subplots with FOR Loop - example #1
Categorical data will be plotted. Each attribute (column name) will have one plot.

a = 2, number of rows
b = 3, number of columns
c = 1, initialize plot counter
subplot group: (2,3,c)

```python
df.columns
```

```python
#  Categorical Data
cars_CAT = ['vs', 'gear', 'cyl', 'carb']

#  Categorical Data
a = 2  # number of rows
b = 3  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize = (8,6))
plt.suptitle("SUBPLOTS with FOR loop - example #1", fontsize = 18)

for i in cars_CAT:
    plt.subplot(a, b, c)
    plt.title('{}, subplot: {},{},{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.countplot(data = df, x = i, hue = 'am')
    c = c + 1

plt.tight_layout()
plt.show()
```

5. Subplots with FOR Loop - example #2
Numerical data will be plotted. Each attribute (column name) will have three plots (distribution, box and scatter).

a = 4, number of rows
b = 3, number of columns
c = 1, initialize plot counter
subplot group: (4,3,c)
'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

```python
#  Numerical Data
cars_NUM = ['mpg', 'disp', 'hp', 'drat', 'wt', 'qsec']
```

```python
#  plot Numerical Data
a = len(cars_NUM)  # number of rows
b = 3  # number of columns
c = 1  # initialize plot counter

fig = plt.figure(figsize = (10,14))
plt.suptitle("SUBPLOTS with FOR loop - example #2", fontsize = 18)

for i in cars_NUM:
    plt.subplot(a, b, c)
    plt.title('{} (dist), subplot: {},{},{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.kdeplot(data = df, x = i, hue ='am')
    c = c + 1

    plt.subplot(a, b, c)
    plt.title('{} (box), subplot: {},{},{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.boxplot(data = df, y = i, x = 'am')
    c = c + 1

    plt.subplot(a, b, c)
    plt.title('{} (scatter), subplot: {},{},{}'.format(i, a, b, c))
    plt.xlabel(i)
    sns.scatterplot(data = df, x = i, y = 'mpg', hue = 'am')
    c = c + 1

plt.tight_layout()
plt.show()
```

6. Heatmaps with Subplots
Correlation heatmaps for both the sexes are plottled side-by-side (1x2 subplot).

1, number of rows
2, number of columns
subplot group: (1,2,c)

```python
# select few attributes
df_corr = df[['am','mpg', 'wt','hp','drat']]

# correlation - female
dfAuto     = df_corr[(df_corr['am'] == 0)]    # df-auto
dfAutoCorr = dfAuto.drop(["am"], axis=1).corr()    # df-autotfemale corr
# correlation - male
dfManual       = df_corr[(df_corr['am'] == 1)]      # df-manual
dfManualCorr   = dfManual.drop(["am"], axis=1).corr()      # Manual corr


#  SUBPLOTS - heatmaps, 1x2
fig = plt.figure(figsize = (8,8))

#  subplot 1 - Auto
plt.subplot(121)
plt.title('correlation - Auto  (1,2,1)', fontsize = 16)
sns.heatmap(dfAutoCorr, annot = True, cbar = False, fmt = '.2f', square = True, cmap = 'Reds_r')

#  subplot 2 - Manual
plt.subplot(122)
plt.title('correlation - MALE  (1,2,2)', fontsize = 16)
sns.heatmap(dfManualCorr, annot = True, cbar = False, fmt = '.2f', square = True, cmap = 'Blues_r')

plt.tight_layout()
plt.show()
```
