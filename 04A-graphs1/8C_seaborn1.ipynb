#### Grid type plot: This example shows a regression plot of tips vs the total_bill from the dataset. lmplot stands for linear model plot and is used to create a regression plot. x =’total_bill’ sets the x axis to total_bill. y=’tip’ sets the y axis to tips. height=2 is used to the size(the height)of the plot. aspect is used to set the width keeping the width constant.

```python
tips = sns.load_dataset('tips')
sns.lmplot(x ='total_bill', y ='tip', height=3, aspect = 4, data = tips)
```

```python
#Scale and Context
sns.set_context('poster', font_scale = 2)
sns.countplot(x ='sex', data = tips, palette ='coolwarm')
```

## Seaborn Grid

```python
# load the tips dataset present by default in seaborn
tips = sns.load_dataset('tips')
sns.set_style('white')
 
# make a countplot
sns.countplot(x ='sex', data = tips)
```

```python
############# Main Section         #############
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, col ="sex",  hue ="day")
# map the above form facetgrid with some attributes
graph.map(plt.scatter, "total_bill", "tip", edgecolor ="w").add_legend()
# show the object
plt.show();
```

```python
## Histogram
```

```python
############# Main Section         #############
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, row ='smoker', col ='time')
# map the above form facetgrid with some attributes
graph.map(plt.hist, 'total_bill', bins = 15, color ='orange')
# show the object
plt.show()
```

```python
#  Regression Plot
# Form a facetgrid using columns with a hue
graph = seaborn.FacetGrid(df, col ='time', hue ='smoker')
# map the above form facetgrid with some attributes
graph.map(seaborn.regplot, "total_bill", "tip").add_legend()
# show the object
plt.show();
  
```
