## sns Box Plot
-  https://seaborn.pydata.org/generated/seaborn.boxplot.html
-  Grouped Box  Plots

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the example tips dataset
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
tips.head()
```

```python
# Draw a nested boxplot to show bills by day and time : G
sns.boxplot(x="day", y="total_bill",   hue="smoker", palette=["m", "g"],
    data=tips)
sns.despine(offset=10, trim=True)
```

```python
sns.set_theme(style="ticks")
# Load the example planets dataset
planets = sns.load_dataset("planets")
```

```python
sns.boxplot(x=titanic["age"])
```

```python
sns.boxplot(data=titanic, x="age", y="class")
```

```python
sns.boxplot(data=titanic, x="class", y="age", hue="alive")
```

```python
# Initialize the figure with a logarithmic x axis
f, ax = plt.subplots(figsize=(7, 6))
ax.set_xscale("log")
# Plot the orbital period with horizontal boxes
sns.boxplot(  planets, x="distance", y="method", hue="method",
    whis=[0, 100], width=.6, palette="vlag" )

# Add in points to show each observation
sns.stripplot(planets, x="distance", y="method", size=4, color=".3")

# Tweak the visual presentation
ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

# Violin Plot
-  https://seaborn.pydata.org/generated/seaborn.violinplot.html

```python
titanic = sns.load_dataset("titanic")
sns.violinplot(x=titanic["age"])
```

```python
sns.violinplot(data=titanic, x="age", y="class")
```

```python
sns.violinplot(data=titanic, x="class", y="age", hue="alive")
```

```python
sns.violinplot(data=titanic, x="class", y="age", hue="alive", fill=False)
```

```python
sns.violinplot(data=titanic, x="class", y="age", hue="alive", 
               split=True, inner="quart")
```

```python
sns.violinplot(data=titanic, x="class", y="age", hue="alive", split=True, gap=.1, inner="quart")
```

```python
sns.violinplot(data=titanic, x="class", y="age", split=True, inner="quart")
```

```python
sns.violinplot(data=titanic, x="age", y="deck", inner="point")
```

```python
sns.violinplot(data=titanic, x="age", y="deck", inner="point", density_norm="count")
```

```python
sns.violinplot(data=titanic, x="age", y="alive", cut=0, inner="stick")
```

```python
sns.violinplot(data=titanic, x="age", y="alive", bw_adjust=.5, inner="stick")
```

```python
sns.violinplot(x=titanic["age"].round(-1) + 5, y=titanic["fare"], native_scale=True)
```
