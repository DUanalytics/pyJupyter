## sns Histogram
-  https://seaborn.pydata.org/examples/three_variable_histogram.html

```python
import seaborn as sns
sns.set_theme(style="dark")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
diamonds = sns.load_dataset("diamonds")
iris = sns.load_dataset("iris")
```

```python
sns.histplot(data=iris, x="sepal_length")
```

```python
sns.histplot(data=iris, x="sepal_length", kde=True)
```

```python
sns.displot(iris["sepal_length"])
```

```python
#controlling bins
sns.histplot(data=iris, x="sepal_length", bins=20)
```

```python
sns.displot(
    data=diamonds, x="price", y="color", col="clarity",
    log_scale=(True, False), col_wrap=4, height=4, aspect=.7,
)
```
