# Regression Plots
-  https://seaborn.pydata.org/examples/multiple_regression.html

```python
import seaborn as sns
sns.set_theme()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the penguins dataset
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Plot sepal width as a function of sepal_length across days
g = sns.lmplot(   data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    height=5)

# Use more informative axis labels than are provided by default
g.set_axis_labels("Snoot length (mm)", "Snoot depth (mm)")
```

## Regression fit over a stip plot
-  https://seaborn.pydata.org/examples/strip_regplot.html

```python
mpg = sns.load_dataset("mpg")
sns.catplot(  data=mpg, x="cylinders", y="acceleration", hue="weight",
    native_scale=True, zorder=1)
sns.regplot(   data=mpg, x="cylinders", y="acceleration",
    scatter=False, truncate=False, order=2, color=".2",)
```
