```python
#Mall Customers Data
colNames = ['customerID','gender','age','annualIncome','spendingScore']
df = pd.read_csv('https://raw.githubusercontent.com/DUanalytics/datasets/master/csv/clusteringMallCustomers.csv', 
                 skiprows = 1, header=None, names= colNames, usecols=['age','spendingScore','annualIncome'], 
                 nrows=10)
df.head()
```

```python
rows = ['R0','R1','R2','R3','R4','R5','R6','R7','R8','R9']
```

```python
df.index = rows
```

```python
df
```

```python
df1 = df.copy()
```

```python
# compute ED between all rows
import itertools
for i, j in itertools.combinations(df1.index,2):
    d_ij = np.linalg.norm(df1.loc[i] - df1.loc[j])
```

```python
d_ij
```

```python
from scipy.spatial.distance import cdist
EDB1 = np.around(cdist(df1, df1, 'euclid'),1)
print(EDB1)
```

```python
from scipy.spatial.distance import pdist
EDB2 = np.around(pdist(df1.values, 'euclid'),1)
print(EDB2)
```

```python
len(pdist(df1.values, 'euclid'))
```

```python
#rows = ['R0','R1','R2','R3','R4','R5','R6','R7','R8','R9']
df2 = pd.DataFrame(EDB1, columns = rows, index = rows)
df2
```
