```python
#standard libaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Data
# Load transactions from pandas
DL = 'https://raw.githubusercontent.com/DUanalytics/pyAnalytics/master/data/groceries.csv'
groceries = pd.read_csv(DL) #,header=None)
groceries
```

header is actually the first transaction. Each row corresponds to a transaction and each column corresponds to an item purchased in that specific transaction. The NaN tells us that the item represented by the column was not purchased in that specific transaction.

```python
#reload data with header None
data = pd.read_csv(DL, names=['products'], sep='|') # header=None,
data
```

```python
data.shape
```

```python
data2 = list(data['products'].apply(lambda x:x.split(",")))
data2
```

```python
#Let's transform the list, with one-hot encoding
from mlxtend.preprocessing import TransactionEncoder
a = TransactionEncoder()
a_data = a.fit(data2).transform(data2)
df = pd.DataFrame(a_data,columns=a.columns_)
#df = df.replace(False, 0)
#df = df.replace(True, 1)
df
```

```python
#set a threshold value for the support value and calculate the support value.
df2 = apriori(df, min_support = 0.001, use_colnames = True, verbose = 1)
df2.head()
```

```python
df2.shape
```

```python
# Support
df2_ar_sp = association_rules(df2, metric = "support", min_threshold = 0.04)
df2_ar_sp.sort_values(by='support', ascending=False).head(10)
```

```python
#Let's view our interpretation values using the Association rule function.
df2_ar_conf = association_rules(df2, metric = "confidence", min_threshold = 0.08)
df2_ar_conf.sort_values(by='confidence', ascending=False).head(10)
```

## Lift

```python
df2_ar_lift = association_rules(df2, metric = "lift", min_threshold = 2.5)
df2_ar_lift.sort_values(by='lift', ascending=False).head(10)
```
