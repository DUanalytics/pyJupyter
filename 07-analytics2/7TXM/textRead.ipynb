## Read from text document file
https://jamesclear.com/great-speeches/what-matters-more-than-your-talents-by-jeff-bezos

```python
raw = open('jeffBoss.txt', encoding="utf8").read()
type(raw)
raw
```

```python
tokens = nltk.word_tokenize(raw)
print(type(tokens), tokens)
```

```python
words = [w.lower() for w in tokens]
print(type(words), ' ::', words)
```

```python
vocab = sorted(set(words))
print(type(vocab), ' :: ', vocab, '\t')
```
