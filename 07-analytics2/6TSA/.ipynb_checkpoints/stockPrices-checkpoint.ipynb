# Time Series - under development
Time series is a series of data points in which each data point is associated with a timestamp. A simple example is the price of a stock in the stock market at different points of time on a given day. Another example is the amount of rainfall in a region at different months of the year.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
# General routine for plotting time series data
def plot_timeseries_df(df, attrib, ticker_loc=1, title='Timeseries',    legend=''):
    fig = plt.figure(figsize=(15,7))
    plt.plot(df[attrib], 'o-')
    _ = plt.xticks(rotation=90)
    plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(ticker_loc))
    plt.title(title)
    plt.gca().legend(legend)
    plt.show()
 
plot_timeseries_df(shares_multiple_df.loc["2022-10-01":"2022-11-01"], "Close",
                   ticker_loc=3, title="Close price", legend=companies)
```
