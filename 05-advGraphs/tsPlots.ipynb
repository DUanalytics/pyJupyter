```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#
tsData = pd.Series(np.random.randn(1000), index = pd.date_range('1/1/2022', periods=1000))
tsData
```

```python
ts1 = tsData.cumsum()
ts1
```

```python
ts1.plot()
```

```python
#data for multi series
tsData2 = pd.DataFrame(np.random.randn(1000,4), index = tsData.index, columns = list("ABCD"))
tsData2
```

```python
ts2 = tsData2.cumsum()
plt.figure()
ts2.plot()
plt.show();
```
