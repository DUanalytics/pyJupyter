```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Data

```python
#sample data 
x = [1,2,3]
y = [2,4,1]
print(x, y)
```

```python
np.column_stack((x,y))
```

## Plot-1 : Line

```python
plt.plot(x,y)  #plt function comes from matplotlib.pyplot
```

## Plot-2 : Line with Legend

```python
#plot with labelling axis and title
plt.plot(x,y, label='Line 1')
plt.xlabel('X Data')
plt.ylabel('Y Data')
plt.title('My First Graph')
plt.show()
```
