## Adjusting Figure Size

```python
#setting figure size
plt.rcParams["figure.figsize"] = [5, 5]   #width, height
# figsize tuple accepts inches, 
#so if you want to set it in centimetres you have to divide them by 2.54
#For those using jupyter notebook, please make sure you are 
#using plt.figure(figsize=(15, 10)) before other plt commands
```

```python
#Reset
plt.rcParams["figure.figsize"] = plt.rcParamsDefault["figure.figsize"]
```

```python
x = [1,2,3]
plt.plot(x, x)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
x = [1,2,3]
plt.plot(x, x)
plt.show()
```
