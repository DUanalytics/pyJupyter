## Tree Maps
https://python-charts.com/part-whole/treemap-matplotlib/
- the squarify library provides a function named squarify.plot which can be used to create beautiful treemaps in Python. In order to create a basic treemap pass an array of values to the sizes argument. Note that this kind of graph doesnâ€™t need an axis, so you can remove it with plt.axis("off").

```python
import matplotlib.pyplot as plt
import squarify  #install 
# extras
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#!pip install squarify
```

```python
#plot1 
# Sample data
values = [250, 120, 280, 320, 140, 95]
# Treemap
squarify.plot(sizes = values)
# Remove the axis:
plt.axis("off")   #note
# plt.show()
```

```python
# Treemap
squarify.plot(sizes = values,  pad = 0.25)
# Remove the axis:
plt.axis("off")
# plt.show()
```

```python
#Border
# Sample data
values = [250, 120, 280, 320, 140, 95]
# Treemap
squarify.plot(sizes = values,  ec = 'black')

# Remove the axis:
plt.axis("off")

# plt.show()
```

```python
#axis scaling Treemap
squarify.plot(sizes = values,    norm_x = 200,   norm_y = 20)

# plt.show()
#By default, treemaps made with squarify are of size 100x100, but with the norm_x and
#norm_y arguments you can override the dimensions if needed of the X and Y-axis, respectively.
```

```python
#color palette
# Sample data
values = [250, 120, 280, 320, 140, 95]
colors = ['#91DCEA', '#64CDCC', '#5FBB68',   '#F9D23C', '#F9A729', '#FD6F30']

squarify.plot(sizes = values, color = colors, alpha = 0.7)

# Remove the axis:
plt.axis("off")

plt.show()
```

```python
# Sample data
values = [250, 120, 280, 320, 140, 95]
# Treemap
squarify.plot(sizes = values,color = sns.color_palette("magma", len(values)), alpha = 0.7)
# Remove the axis:
plt.axis("off")
plt.show()
```

```python
# Labels
# Sample data
values = [250, 120, 280, 320, 140, 95]
labels = ['Group 1\n 250', 'Group 2\n 120', 'Group 3\n 280',  'Group 4\n 320', 'Group 5\n 140', 'Group 6\n 95']
colors = ['#91DCEA', '#64CDCC', '#5FBB68',  '#F9D23C', '#F9A729', '#FD6F30']
# Treemap
squarify.plot(sizes = values, label = labels,  color = colors, alpha = 0.7)
# Remove the axis:
plt.axis("off")
plt.show();
```
