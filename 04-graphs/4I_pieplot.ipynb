# Pie Chart

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#sample Data
course = ['Diploma', 'BBA', 'MBA', 'PHD']
students = np.random.randint(100,150, size=4)
df1 = pd.DataFrame(list(zip(course, students)), columns=['course', 'students'])
df1
```

```python
#colors
colors1 = ['r','y','b','g']
```

```python
#plot pie
plt.pie(students, labels=course)
plt.legend()
#plt.show()  #this will show less comments on output
```

```python
plt.pie(students, labels=course, colors=colors1, startangle=90, shadow=True, explode=[0,0,0,.1]
       , radius=2, autopct ='%1.1f%%')
plt.legend()
plt.show()
```

```python
#comments
#define the labels using a list called activities.
#Each label can be defined using another list called courses.
#Color for each label is defined using a list called colors1.
#shadow = True will show a shadow beneath each label in pie chart.
#startangle rotates the start of the pie chart by given degrees counterclockwise from the x-axis.
#explode is used to set the fraction of radius with which we offset each wedge.
#autopct is used to format the value of each label. Here, we have set it to show the percentage value only upto 1 decimal place.
```
