```python
# dictionary of lists
dict = {'subject1':[np.nan, 100, 90, np.nan, 95],
        'subject2':[np.nan, 30, 45, 56, np.nan],
        'subject3':[np.nan, 45, 40, 80, 98],
        'subject4':[np.nan, np.nan, np.nan, np.nan, 65],
        'subject5':[np.nan, np.nan, np.nan, np.nan,np.nan]}
 
# creating a dataframe from list
df = pd.DataFrame(dict)
df1 = df.copy()
```

```python
#interpolate
df1.interpolate(method ='linear', limit_direction ='forward')
#values in the first row could not get filled as the direction of filling of values is forward and there is no previous value which could have been used in interpolation. 
```

```python
#drop rows with at least 1 null value
df1.dropna(axis=0)
```

```python
df1.dropna(axis=1)
```

```python
# drop rows which have all data missing 
df1.dropna(how = 'all') #remove row which is totally blank
```

```python
df1.dropna(how = 'all',axis=1)  #remove columns which totally blank
```

```python
df1.dropna(axis = 0, how ='any')  #any missing in row
```

```python
df1.dropna(axis = 1, how ='any')  #any missing in column
```

```python
df1.dropna(axis = 1, thresh=2)  #min 1 null value in a column
```

```python
df1.dropna(axis = 0, thresh=2)  # min 2 null value in a row
```
