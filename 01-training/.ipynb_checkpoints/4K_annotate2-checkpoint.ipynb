```python
#import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#clear any plot in the region
plt.clf()
```

```python
x = np.arange(0,10,1)
y = np.round(np.random.normal(loc=5, scale=1, size=10),2)
np.column_stack((x,y))
```

```python
plt.plot(x,y)
plt.show();
```

```python
plt.plot(x,y)
plt.text(2, 5, "This is text at 1 point")
plt.show();
```

```python
plt.plot(x,y)
plt.text(2, 5, "This is text at 1 point with Horz Alighment", horizontalalignment ='right')
plt.show();
```

```python
plt.plot(x, y, 'bo-')
for x,y in zip(x, y):
    label = '{:.2f}'.format(y)
    plt.annotate(label, (x,y), textcoords='offset points', xytext=(0,10), ha='center')
plt.show();
```

```python
#bar plot  - is also similar
x1 = np.arange(0,5,1)
y1 = np.random.randint(10, 20, size=5)
x1,y1
plt.bar(x1, y1)
for x,y in zip(x1, y1):
    label = '{:.2f}'.format(y)
    plt.annotate(label, (x,y), textcoords='offset points', xytext=(0,10), ha='center')
plt.show();
```
