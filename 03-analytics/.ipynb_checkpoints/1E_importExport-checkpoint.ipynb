```python
#Topic: Import data from CSV, Excel, URL and Gsheet
import pandas as pd
#link https://docs.google.com/spreadsheets/d/1X2AjCyh1Kcs-CA8ilGZyBZBobuFP7cSS9D-3a5LK6sk/edit#gid=1858159025
student = pd.read_csv('https://docs.google.com/spreadsheets/d/1X2AjCyh1Kcs-CA8ilGZyBZBobuFP7cSS9D-3a5LK6sk' 
                      + '/export?gid=1858159025&format=csv', index_col=0 )
#sheet public shareable
student.head(2)
```

```python
#pandas melt - wide to long
#https://pandas.pydata.org/docs/reference/api/pandas.melt.html
pd.melt(student, id_vars=['sname'], value_vars=['class10','class12'],
        var_name='schoolLevel', value_name='marks')
```

```python
pd.melt(student, id_vars=['sname'], value_vars=['class10','class12'],
        var_name='schoolLevel', value_name='marks', ignore_index=False)
```

```python
#CSV online File
#link https://raw.githubusercontent.com/DUanalytics/datasets/master/csv/salesdata.csv
#help(pd.read_csv)
sales = pd.DataFrame(pd.read_csv(
    'https://raw.githubusercontent.com/DUanalytics/datasets/master/csv/salesdata.csv', 
    index_col=None))
sales.head(2)
```

```python
sales2 = sales.drop(columns=['Unnamed: 0'], axis=1)  #remove first column
sales2.head(5)
```

```python
sales2.reset_index()
sales2.columns
```

```python
sales2.describe()
```

```python
pd.melt(sales2, id_vars=['month'], value_vars=['sales'])
#help(pd.pivot)
```

```python
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.pivot_table.html
sales2
```

```python
pd.pivot_table(sales2, values='sales', columns=['month'], aggfunc=np.sum)
```

```python
pd.pivot(data=sales2, index='month', columns=['productID'], values='sales')
```

```python
sales3 = pd.concat([sales2, sales2])
sales3.head()
```
