```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data
csvLink ='https://raw.githubusercontent.com/DUanalytics/datasets/master/csv/studentdata.csv'
df = pd.DataFrame(pd.read_csv(csvLink, index_col=None))
#studentData = pd.DataFrame(pd.read_csv(csvLink))
df.head(5)
```

```python
#wide to long
df_melted = pd.melt(df, id_vars=['rollno'], value_vars=['EXCEL','MATHS', 'RPGM', 'STATS'],
        var_name='subject', value_name='marks', ignore_index=False)
df_melted
```

```python
df_melted.subject.value_counts()
```
